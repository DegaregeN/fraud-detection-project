```python
# Removing duplicates
df = df.drop_duplicates()
```

```python
# Plot histogram for 'Amount'
plt.figure(figsize=(10, 6))

# Plotting the histogram with more bins
sns.histplot(df['Amount'], bins=6, kde=True, color='skyblue')

# Calculating the mean and median
mean_amount = np.mean(df['Amount'])
median_amount = np.median(df['Amount'])

# Adding lines for mean and median
plt.axvline(mean_amount, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_amount:.2f}')
plt.axvline(median_amount, color='green', linestyle='--', linewidth=2, label=f'Median: {median_amount:.2f}')

# Adding titles and labels
plt.title('Distribution of Transaction Amounts', fontsize=16)
plt.xlabel('Transaction Amount (USD)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

```python
# Bivariate analysis
import seaborn as sns
sns.boxplot(x='Class', y='Amount', data=df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram for 'Time'
plt.figure(figsize=(10, 6))
sns.histplot(df['Time'], bins=50, kde=False, color='blue')

# Adding titles and labels
plt.title('Distribution of Transaction Time', fontsize=16)
plt.xlabel('Time (seconds since first transaction)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count plot for the 'Class' column
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='Class', data=df, palette='pastel')

# Adding counts on top of the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', fontsize=12)

# Adding titles and labels
plt.title('Distribution of Class (Fraud vs Non-Fraud)', fontsize=16)
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Display the plot
plt.show()
```
